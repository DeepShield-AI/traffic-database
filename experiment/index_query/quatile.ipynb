```python
import numpy as np
```

```python
with open("../new_index_query/iscx_range.log") as file:
    lines = file.readlines()

data_dict = {}

id = -1

for line in lines:
    if "ID" in line:
        id += 1
        data_dict[id] = []
        continue
    if id >=0:
        data_dict[id].append(int(line.split(' ')[0]))

data_dict
```

```python
quatile_dict = {}

for id,data in data_dict.items():
    quatile_dict[id] = {}
    for i in range(0,11):
        quatile_dict[id][i*10] = np.percentile(data,i*10)

quatile_dict
```

```python
for id, q in quatile_dict.items():
    print(id)
    for k,v in q.items():
        print(v/1000)
```

```python
with open("../new_index_query/iscx_combined.log") as file:
    lines = file.readlines()

data_dict = {}
id = -1

for line in lines:
    if "ipv4" in line or "ipv6" in line:
        id += 1
        data_dict[id] = []
        continue
    if id >=0:
        data_dict[id].append(int(line))

data_dict
```

```python
quatile_dict = {}

for id,data in data_dict.items():
    quatile_dict[id] = {}
    for i in range(0,11):
        quatile_dict[id][i*10] = np.percentile(data,i*10)

for id, q in quatile_dict.items():
    print(id)
    for k,v in q.items():
        print(v/1000)
```
